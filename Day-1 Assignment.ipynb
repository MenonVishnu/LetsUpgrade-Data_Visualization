```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Question 1:
plot a line plot between a and b:<br>
a=np.arange(40,50)<br>
b=np.arange(50,60)

```python
a = np.arange(40,50)
b = np.arange(50,60)

plt.title("Question 1:", color='Blue')
plt.xlabel("X-Axis", color='Red')
plt.ylabel("Y-Axis", color='Red')
plt.plot(a,b,'g*-',linewidth=1,markersize=14)
plt.show()
```

## Question 2:
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week<br>
sales_1 = [160,150,140,145,175,165,180] #sales of company1<br>
sales_2 = [70,90,160,150,140,145,175]  #sales of company2<br>


```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180] #company1
sales_2 = [70,90,160,150,140,145,175] #company2

plt.title("Question 2: ",color='blue')
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')

plt.plot(days, sales_1, 'ro-', markersize=12, linewidth=2)
plt.plot(days, sales_2, 'go-', markersize=12, linewidth=2)
plt.show()
```

## Question 3:
Create a 3 by 3 subplots:<br>
multiple plots<br>
x = [1,2,3,4]<br>
y1 = [4,3,2,1]<br>
y2 = [10,20,30,40]<br>
y3 = [40,30,20,10]<br>
y4 = [1,2,1,2]<br>
y5 = [40,70,90,70]<br>


```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplots(figsize=(20,20))

plt.subplot(3,2,1)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y1,'go--')

plt.subplot(3,2,2)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y2,'y*-')

plt.subplot(3,2,3)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y3,'r+--', markersize=10)

plt.subplot(3,2,4)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y4,'bo-')

plt.subplot(3,2,5)
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')
plt.plot(x,y5,'g*-')

plt.show()
```
